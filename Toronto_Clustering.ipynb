```python
import pandas as pd
link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df = pd.read_html(link,header=0)[0]
df
```

```python
df = df[~(df['Borough']=='Not assigned')]
df.head()
```

```python
df.reset_index(inplace=True)
df.drop(columns='index',axis=1,inplace=True)
df
```
